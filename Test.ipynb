## Libraries

```python
from tensorflow import keras

from os import listdir
from os.path import isfile, join

import pandas as pd
import numpy as np
from PIL import Image
```

## Load Images

```python
folder_path = "/content/drive/MyDrive/Enzo_2023/ISIMA/Deep Learning/TP noté/images"
files = [f for f in listdir(folder_path) if isfile(join(folder_path, f))]

df = pd.DataFrame(columns=['image_name', 'relative_path', 'class_label'])
df['image_name'] = files

for i in range(len(files)):
  df.loc[i, 'relative_path'] = folder_path + "/" + files[i]
```

```python
print(df.head())
```

## Load Model

```python
model_path = "./final_model.h5"
model = keras.models.load_model(model_path)
```

## Make Predictions

```python
x_test = []
for i in range(len(df)):
    try:
        train_path = df.loc[i, 'relative_path']
        image = Image.open(train_path)
        images_arr = np.asarray(image)
        x_test.append(images_arr)
    except:
        print("ERREUR : Fichier train PNG non trouvé.")
        break
x_test = np.array(x_test).reshape(-1, 28, 28, 1)
```

```python
pred = model.predict(x_test)
y_pred_cp = []

for i in range(pred.shape[0]):
    y_pred_cp.append(np.argmax(pred[i]))

y_pred_cp = np.array(y_pred_cp)

print(f"shape y_pred_cp  = {y_pred_cp.shape}")
```

```python
df['class_label'] = y_pred_cp
```

## Save Predictions

```python
df.to_csv('./prediction.csv')
```
